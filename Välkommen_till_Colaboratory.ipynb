  <h2>Vad är Colab?</h2>
</div>

I Colab, eller Colaboratory, kan du skriva och köra Python i webbläsaren med följande fördelar:
- Ingen konfigurering krävs
- Få åtkomst till grafikprocessorer kostnadsfritt
- Enkel delning

Oavsett om du är <strong>student</strong>, <strong>datavetare</strong> eller <strong>forskare inom AI </strong> kan Colab göra ditt arbete enklare. Titta på <a href="https://www.youtube.com/watch?v=inN8seMm7UI">Introduktion till Colab</a> om du vill meta mer, eller kom igång nedan.

## <strong>Komma igång</strong>
</div>

Dokumentet du läser är inte en statisk webbplats utan en interaktiv miljö som kallas för <strong>Colab-anteckningsbok</strong>. I den kan du skriva och köra kod.

Exempel: Här visas en <strong>kodcell</strong> med ett kort Python-skript som beräknar ett värde, lagrar det i en variabel och skriver ut resultatet:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

Om du vill köra kod i cellen ovan markerar du den med ett klick och trycker sedan på knappen Spela upp till vänster om koden eller använder kortkommadot Cmd/Ctrl+Enter. Om du vill redigera koden klickar du på cellen och börjar redigera.

Variabler som du definierar i en cell kan användas i andra celler senare:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Med Colab-anteckningsböcker kan du kombinera <strong>körbar kod</strong> och <strong>rich text</strong> i ett enda dokument tillsammans med <strong>bilder</strong>, <strong>HTML</strong>, <strong>LaTeX</strong> och mycket annat. När du skapar egna Colab-anteckningsböcker sparas de i ditt Google Drive-konto. Du kan enkelt dela Colab-anteckningsböckerna med kollegor eller vänner och låta dem skriva kommentarer i anteckningsböckerna eller redigera dem. Läs <a href="/notebooks/basic_features_overview.ipynb">Översikt över Colab</a> om du vill veta mer. Du skapar en ny Colab-anteckningsbok via menyn Arkiv ovan eller via följande länk: <a href="http://colab.research.google.com#create=true">Skapa en ny Colab-anteckningsbok</a>.

Colab-anteckningsböcker är Jupyter-anteckningsböcker som Colab värdlagrar. Du kan läsa mer om Jupyter-projektet på <a href="https://www.jupyter.org">jupyter.org</a>.

## Datavetenskap
</div>

Med Colab kan du dra nytta av kraften hos Python-bibliotek för att analysera och visualisera data. I kodcellen nedan används <strong>numpy</strong> för att generera slumpmässig data och <strong>matplotlib</strong> för att visualisera den. Om du vill redigera koden klickar du på cellen och börjar redigera.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

Du kan importera egen data i Colab-anteckningsböcker från Google Drive-kontot, t.ex. från kalkylark, GitHub och flera andra källor. I länkarna under <a href="#working-with-data">Arbeta med data</a> kan du läsa mer om hur du importerar data och hur Colab kan användas för datavetenskap.
